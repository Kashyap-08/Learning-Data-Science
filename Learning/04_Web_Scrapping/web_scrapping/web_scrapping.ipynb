```python
# parse image
image = BeautifulSoup(response.content, 'html.parser')
```

```python
# select content where tags are img
image_tags = image.find_all('img')
```

```python
for i in image_tags:
    # select tags where named as src. 'src' contain image link
    image_url = i['src']

    # extract image form web. extracted images is in byte format.
    byte_image = requests.get(image_url).content

    # open the folder where we want to store images and give image name.
    with open(os.path.join(save_dir, f"{query}_{image_tags.index(i)}.jpg"), 'wb') as f:
        f.write(byte_image)
        
        
```

## Create a new method that returns the saved image just by passing path and image name

```python
def download_images(query,direc):
    save_dir = f"{direc}/"
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)

    url = f"https://www.google.com/search?q={query}&tbm=isch&ved=2ahUKEwiAsJebhbiAAxVE3DgGHcNrBZYQ2-cCegQIABAA&oq=elon+m&gs_lcp=CgNpbWcQARgAMgoIABCKBRCxAxBDMggIABCABBCxAzIKCAAQigUQsQMQQzIKCAAQigUQsQMQQzIICAAQgAQQsQMyCAgAEIAEELEDMggIABCABBCxAzIICAAQgAQQsQMyCAgAEIAEELEDMggIABCABBCxAzoECCMQJzoECAAQAzoLCAAQgAQQsQMQgwE6BQgAEIAEOggIABCxAxCDAToHCAAQigUQQ1DhCli3FWCYIWgAcAB4AIAB1wGIAZAKkgEFMC42LjGYAQCgAQGqAQtnd3Mtd2l6LWltZ8ABAQ&sclient=img&ei=bC_HZIDKMMS44-EPw9eVsAk&bih=739&biw=1536&rlz=1C1YTUH_enIN1061IN1061"
    response = requests.get(url)
    images = BeautifulSoup(response.content, 'html.parser')
    image_tagss = images.find_all('img')
    del image_tagss[0]
    for i in image_tagss:
        image_url = i['src']
        byte_image = requests.get(image_url).content
        with open(os.path.join(save_dir,f"{query}_{image_tagss.index(i)}.jpg"),'wb') as f:
            f.write(byte_image)
    
```
