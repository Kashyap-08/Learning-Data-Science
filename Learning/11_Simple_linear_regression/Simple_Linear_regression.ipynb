## 3. Bias and Varience


### Overfitting: 
* Training data -> Hight accuract (**Low Bias**)
* Test data -> Low Accuracy (**High Varience**)

* **Bias:** In this context, bias is generally low because model can fit the training data quiete well.
* **Variance:** High Varience is probably when your models prediction are sensative to small changes in training data, which can lead to poor generalization and low performance on new data.


#### To address this issue, you might consider techniques to reduce overfitting, such as:

* Regularization: Introduce penalties on model complexity to discourage overfitting (e.g., L1 or L2 regularization).

* More Data: Collect additional data to provide the model with a broader and more representative sample of the underlying data distribution.

* Feature Selection: Carefully select relevant features and remove irrelevant or noisy features.

* Cross-Validation: Use cross-validation techniques to assess model performance more reliably and to find a better balance between bias and variance.

* Simpler Models: Consider using simpler model architectures or ensembling methods to reduce model complexity.



### Underfitting:
* Training data -> Low accuract (**High Bias**)
* Test data -> High Accuracy (**Low Varience**)

* **Bias:** In this context, bias is generally high because model is unable to fir the training data effectively, indicating a substantial error.
* **Variance:** Varience is low because the model is simple and does not exhabits the sensitivity to noise or small fluctuations in the training data.


* =========================== OR ===========================

* Training data -> Low accuract (**High Bias**)
* Test data -> Low Accuracy (**High Varience**)

* High Bias: The low training score suggests that the model is too simplistic or not complex enough to capture the underlying patterns in the training data. In other words, it has a high bias because it cannot represent the data adequately.

* High Variance: The low test score indicates that the model is not generalizing well to unseen data, suggesting that it exhibits high variance. Even though it's not overfitting the training data, it's still unable to capture the underlying data distribution effectively.
    

#### To address this issue and improve model performance, you might consider the following:

* Model Complexity: Try increasing the model's complexity by using a more powerful architecture, adding more layers or neurons (in the case of neural networks), or using a more expressive machine learning algorithm.

* Feature Engineering: Ensure that your feature selection and engineering processes are capturing the relevant patterns in the data.

* Hyperparameter Tuning: Experiment with different hyperparameters to fine-tune the model's performance. This may include adjusting learning rates, regularization parameters, or other model-specific settings.

* More Data: If possible, gather more training data to provide the model with a richer and more representative dataset.

* Ensemble Methods: Consider using ensemble techniques to combine multiple simple models to create a more complex and accurate model.


### Perfect Model (Will-Fitted):
* Training data -> High accuract (**Low Bias**)
* Test data -> High Accuracy (**Low Varience**)

* Low Bias: The high training score indicates that the model fits the training data well and captures the underlying patterns in the data effectively. It doesn't make strong assumptions that lead to systematic errors.

* Low Variance: The high test score suggests that the model generalizes well to new, unseen data. It doesn't exhibit excessive sensitivity to variations or noise in the training data, indicating low variance.

# Simple Linear Regression


* It is a statistical model used in ML and stats to find the relationship between 2 variables. (dependent and independent)
* The goal is to find the linear equation that best describes the relationship between 2 variables.
* The linear regression model (1 indipendent and 1 dependent variable) can be expressed as: **Y = β0 + β1X**
* The linear regression model (Multiple indipendent and 1 dependent variable) can be expressed as: **Y = β0 + β1X1 + β2X2 + β3X3 + ...**
    * Here, Y: dependent Variable.
    * X: Independent variable.
    * β0: y-intercept (Predicted value of Y when X=0)
    * β1: is the slope coefficient for X, which represent the change in Y for one-unit change in X.  


* Ex: If my β1 = 0.5, β2 = 0.6, and β3 = 1.0 and if my output (intersept) moves by one line, then 
    * my X1 will move by 0.5, X2 will move by 0.6 and X3 will move by 1.0

* The main goal is to find the **Best fit Line** in which we have minimum ERROR.
* Best Fit line is used to predict the dependent varable dependent on independent variable.

### Perfromance Metrics

* They are used to evaluate and quantify the quality of a models predictions or the performance of an algorithm.
* They measures how well a model is doing on a specific task or problem.
* Common performace metrics varies depends upon type of problem.
* For Classification problems you might use:   
    * Accuracy, precision, recall, F1-Score, ROC AUC
* For Regression problem metric include:
    * Mean Suared error (MSE), Mean Absolute Error(MAE) and R-Square.

#### 1. R-Squared

* Is a statistical metrci used to evaluate goodness of fit of a regression model, perticularly in the context of linear regression.
* R-square quantifies how well teh model explains the variabiliy in the data.
* It ranges from 0 to 1, while highes value indicate better fit,
    * R-squared = 0: THe model does not explain ant of the variability in the data and it is a poor fit.
    * R-Squared = 1: THe model perfectly explain all of the variability in the data, and it is an excellent fit.
* FOrmula: **R^2 = 1 - Sum of Squares of Residuals (SSE)/Total Sum of Squares (SST)**
* SSE (Sum of Squares of Residuals) represent the sum of squared difference between observed values(Y) and the Predicted value (Ŷ) from the model.
* SST (Total Sum of Squares) represent the sum of the squared differences between the observed values (Y) and the mean of observed values (Y-Bar).

#### 2. Absolute R-Squared:
* Its a modified version of R-Squared (coefficient of determination) metric used in the context of multiple linear regression.
* While R-square measures the proportion of the variance in the dependent variable that is explained by the independet variable.
* IT provides the more accurate assessment of the models goodness of fit, particularly when multiple independent variables are involved.
* formual: 1 - (1 - R^2) n-1 / n- p - 1
    * where: R^2 = Regular R-square value
    * n = number of data points or observation.
    * p - number of independent variables in model.

#### 3. Mean Aquared Error (MSE):
* calculate the average of the squared differences between predicted and actual values.  
* Lower MSE value indicate better model performance.  
* **Formula:** 1/n(∑(i=0)(n) Yi - Y^)^2  OR 1/n(∑(i=0)(n) Yi - ho(X))^2  
* Disadvantages: 
    * Sensative to Outliers
    * Errors are not in the same unit

* Advantages:
    * Equation is differentiable
    * Has only one local or global minima

#### 4. Mean Absolute Error (MAE)
* measures the average of the absolute difference between predicted and actual value.
* **Formula:** 1/n(∑(i=0)(n) Yi - Y^)  OR 1/n(∑(i=0)(n) Yi - ho(X))
* Advantages:
    * Robust to outliers. (since we are not squaring the difference)
    * Errors will be in the same unit

* Disadvantage:
    * Convergence take more time.(Devides the Gradient into Subgradinet and then find the derivative)

![Alt text](images/image10.png)

#### 5. Root Mean Suared Error (RMSE):
* It is the square root of MSE, providing a measure of the average magnitude of errors in the same units as the target variable.

* **Formula:** sqrt(1/n(∑(i=0)(n) Yi - Y^)^2  OR 1/n(∑(i=0)(n) Yi - ho(X))^2)
* Advantage:
    * In the same unit.
    * Penalizes large error

* Disadvantage:
    * Sensative to outliers
    * Lack of Robustness

# Variants of Linear Regression Model

* **Ridge, Lasso, ElasticNet** are the three popular variants of linear regression model, known as **regularized regression techniques**
* They introduce regularization terms to the linear regression model to prevent overfitting and improve model performance.

## Ridge Regression (Reducr Overfitting) - L2 Regularization

* It is used to prevent multicollinearity (high correlations between independent variable) and to reduce the impact of irrelevent variables in a linear regression model.
* It adds an L2 regularization term to linear regression loss function.
* This term is the sum of the squares of the model's coefficients and is scaled by a hyperparameter, often denoted as "λ" (lambda).
* The main goal is to minimize the sum of squared errors (similar to linear regression), while also minimize sumof squared coefficients 
(regularization term)
* **Formula:** Cost Function: [1/n (∑(i=0)(n) (Yi - ho(X))^2)] + λ * (∑(i=0)(n) (θXi)^2)
    * here: λ = Hyperparameter
    * θXi = coefficient of independent variable.
* ![Alt text](images/image12.png)
* Advantage:
    * Can handle muticollinearity and stabilize model by constraining coefficients, which may lead to better generalization to new data.

#### Comparison to Linear Regression

* In linear regression if i get a proper train data, and my model is trained on it. i can get a perfect best fit line with "0" Error.
* But if we give a test data that may contain some outliers, in this case the model will fail.
* This lead to overfitting of model.

## Lasso Regularization (Reduce Overfitting, Feature Selection) - L1 Regularization

* Stands for Lasso regression, short for "Least Absolute Shrinkage and Selection Operator"
* It is used to prevent overfitting and perfrom feature selection.
* It introduces a penalty term to the linear regression loss function to encourage sparsity in the models coefficients.
* This mean, it drives some of the coefficients to Zero, effectively excludes the features from the model
* **Formula:** Cost Function: [1/n (∑(i=0)(n) (Yi - ho(X))^2)] + λ * |∑(i=0)(n) (θXi)|

## Interview Question: What is L1 and L2 Regularization, What is the difference between them?

* This works by adding penalty term to the loss function that the model is trying to minimze.
* This penaty term encourages the model to have smaller coefficient for the features, which in tern, can lead to simpler models and help avoide overfitting.
