```python
# Import lib
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline # we write this line to avoide plt.show(). 
```

```python
# Create array of same length
array1 =  np.arange(1, 10)
array2 = np.arange(11,20)
```

```python
array1,array2
```

Example1:

```python
plt.scatter(array1, array2, c = 'r') # here 'c' stand for color. we have gave collor as red. c='r'
plt.xlabel('x_axis')    # Title for X-  axis
plt.ylabel('y_label')   # Title for Y-Axis
plt.title('2D Scattered plot') # Title for Graph
plt.savefig('First_Scatter.JPG') # Save Graph
```

Example 2: Line Plot

```python
array3 = array1 * array1 # Create new array that is the square of array1
array3
```

```python
plt.plot(array1, array3, c='g')  # Create a line plot 
```

```python
# OR we can try some other type 
plt.plot(array1, array3, 'g*', linestyle='dashed') 
# Insted of getting the chat in lines we tried to get it in point.
```

Example 3: SubPlot

```python
plt.subplot(2,2,1)
plt.plot(array1, array3, 'r--', )

plt.subplot(2,2,2)
plt.plot(array3, array1, 'go', linestyle = 'dashed')

plt.subplot(2,2,3)
plt.plot(array3[::-1], array1, 'b*')

plt.subplot(2,2,4)
plt.plot(array1[::-1], array3[::-1])

# Task: Give x-axis and y-axis name to all graphs
```
