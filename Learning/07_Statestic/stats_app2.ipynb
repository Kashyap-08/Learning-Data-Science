#### Create normal distribution data

```python
mean = 0.5
standard_deviation = 0.2
normal_dist_data = np.random.normal(mean, standard_deviation, 1000)

#plot data
sns.histplot(normal_dist_data, kde=True)
```

#### Create Log Normal distribution data

```python
mean = 3.0
standard_deviation = 1.0
lognormal_data = np.random.lognormal(mean, standard_deviation, 1000)

# Here the data is lognormal distribution and the data is right-skewed.
# right-skewed data and lognormal distribution is same.
```

```python
# visualize data
sns.histplot(lognormal_data, kde=True)
```

```python
# convert lognormal distribution to normal distribution
normal_data = np.log(lognormal_data)

sns.histplot(normal_data, kde=True)
```

#### QQ Plot

```python
plt.figure(figsize=(10,8))

# Plot the lognormal distributed data
plt.subplot(2,2,1)
sns.histplot(lognormal_data, kde=True)

# Visualize lognormla distributed data using probplot
plt.subplot(2,2,2)
stat.probplot(lognormal_data, dist="norm", plot=pylab)

# Plot normla distributed data, that was converted from lognormal data
plt.subplot(2,2,3)
sns.histplot(normal_data, kde=True)

plt.subplot(2,2,4)
stat.probplot(normal_data, dist="norm", plot=pylab)

plt.show()
```
