# Central Limit Theoram

Suppose you have a population with any shape of probability distribution **(Not Necessarily Normal)**, and you draw random sample **(at least 30)** of a fized size from the population.  
As the sample size increases **(It might be anything and take as much as data)**, the distribution of the sample mean approaches a normal distribution, regardless of the shape of the original population distribution.

```python
# Create population data, binomial data
population_data = np.random.binomial(10, 0.5, 10000)
population_mean = np.mean(population_data)
population_median = np.median(population_data)
population_mode = statistics.mode(population_data)

print(f"Population Mean: {population_mean},\n Population Medina: {population_median}, \n Population Mode: {population_mode}")
```

```python
sns.histplot(population_data,kde=True)
```

```python
# Get sample data from the population
sampel_binomial_data = np.random.choice(population_data, 30)
sample_mean = np.mean(sampel_binomial_data)
sample_median = np.median(sampel_binomial_data)
sample_mode = statistics.mode(sampel_binomial_data)
print(f"Sample Mean: {sample_mean}, \n Sample Median: {sample_median}, \n sample mode: {sample_mode}")
```

```python
sns.histplot(sampel_binomial_data, kde=True)
```

# Z-Score

```python
sample_mean = 110
population_mean = 100
population_std = 15
sample_size = 50
alpha = 0.05
confidence_interval = 1-alpha
```

```python
z_score = (sample_mean - population_mean)/(population_std/ np.sqrt(sample_size))
```

```python
z_score
```

```python
z_critical = stats.norm.ppf(confidence_interval)
```

```python
if z_score > z_critical:
    print("Reject Null Hypothesis")
else:
    print("Failed to reject null hypothesis")
```

```python
p_value = 1 - stats.norm.cdf(z_score)
```

```python
if p_value < alpha:
    print("Reject Null Hypothesis")
else:
    print("Failed to reject null hypothesis")    
```

## 2. Two Sample T-Test

```python
# Create random data
virat_score = np.random.randint(1,100, 20)
rohit_score = np.random.randint(1,100,20)
```

```python
virat_mean = np.mean(virat_score)
rohit_mean = np.mean(rohit_score)
print('Virath mean: ', virat_mean, " Rohit Mean: ", rohit_mean)
```

```python
# Define Hypothesis:
# H0: Both sample mean is equal
# H1: Means are different
```

```python
# calculate t-test 
t_value = stats.ttest_ind(virat_score, rohit_score)
p_value = t_value[1]
```

```python
if p_value < alpha:
    print("Reject Null Hypothesis")
else:
    print("failed to reject null hypothesis")
```

## 3. Paired Sample T-Test

```python
# Get the population data
population_data = np.random.randint(10,80,50)
```

```python
first_inning = np.random.choice(population_data, 20)
second_inning = np.random.choice(population_data, 20)
```

```python
first_inning_mean = np.mean(first_inning)
second_inning_mean = np.mean(second_inning)

print("First Inning:",first_inning_mean, "Second Inning Mean:", second_inning_mean)
```

```python
# Get p-value 
stats_val = stats.ttest_rel(first_inning, second_inning)

p_value = stats_val[1]

print(p_value)
```

```python
# Define hypothesis:
# H0: Both the samples are same 
# H1: Both the samples are different
```

```python
if p_value <= 0.05:
    print("Reject Null hypothesis")
else:
    print("Failed to reject Null Hypothesis")
```

Q: Weight of 25 people were taken before enolling then into the nutrition camp. The populatio mean weight is found to be 45kg before starting the camp. After finishing the camp, for the same 25 people, the sample mean was found to be 75 with a standard deviation of 25, Did the fitness camp work?

```python
# Define Hypothesis
Ho = "mu=45 (sample mean is a population mean)"
H1 = "mu != 45 (Sample mean is not  population mean)"
```

```python
population_mean = 45
sample_mean = 75
sample_sd = 25
n = 25
```

```python
t_test = (sample_mean - population_mean) / (sample_sd/np.sqrt(n))
print(t_test)
```

```python
degree_of_freedom = n-1
alpha = 0.05
confidence_interval = 0.95
```

```python
critical_value = stats.t.ppf(confidence_interval, degree_of_freedom)
print(critical_value)
```

```python
if t_test <= critical_value:
    print("Reject null hypothesis")
else:
    print("failed to reject null hypothesis")
```

```python
# Find using p_value
p_value = 1 - stats.t.cdf(t_test, degree_of_freedom)
print(p_value)
```

# Chi-Square Test

Test is perfromed on categorical data.

```python
# load data from seaborn
data = sns.load_dataset('tips')
```

```python
data[["sex","smoker"]]
```

```python
smoker_data = pd.crosstab(data['sex'], data['smoker'])
```

```python
smokers_array = smoker_data.values # Observed data
```

```python
stats_test, p_value, dof, exp_val = stats.chi2_contingency(smokers_array)
```

```python
# Perfrom chi-square test
shape_of_data = smoker_data.shape
degree_of_freedom = (shape_of_data[0] - 1) * (shape_of_data[1] - 1)
degree_of_freedom
```

```python
chi_square_val = sum([(observed - expected)**2 / expected for observed, expected in zip(smokers_array, exp_val)])
```

```python
alpha = 0.05
confidence_interval = 0.95
```

```python
critical_value = stats.chi2.ppf(confidence_interval, dof)
```

```python
if critical_value <= sum([chi_square_val[0], chi_square_val[1]]):
    print('Reject Null hypothesis, there is relation betwen two categorical variables')
else:
    print("Failed to reject null hypothesis, there is no relation between 2 categorical variables")
```
