#### Interview Question: there is 20 to 30% missing data, how do you want to handle that?
* Understanding data:
    - Before Filling out any missing vaue, its important to understand data.
* There are several methods to hande missing values
    - Imputation (Filling missing values with mean, median ,mode or other estimate)
    - Deletion (Remove row or column with missing data)
    - Regression imputation (Use Regression model to predict missing values)
* This approach depend opon the amount and nature of missing data, as well as the goal of the analysis.

### 3. MultiVarient Analysis

```python
sns.scatterplot(
    x=tips['total_bill'],   # data on X-Axis
    y=tips['tip'],          # Data on Y-Axis
    hue=tips['sex'],        # can apply categories
    style=tips['smoker'], 
    size = tips['size']
    )
```

```python
# Visualize data 
sns.barplot(
    x=titanic['Pclass'], 
    y=titanic['Age'],
    hue=titanic['Sex']
    )
```

```python
# box plot
sns.boxplot(x=titanic['Sex'], y=titanic['Age'], hue=titanic['Survived'])
```

```python
# find maximum age of pasenger who didnt servieved
max_age = titanic[titanic['Survived']==0]['Age'].max()
min_age = titanic[titanic['Survived']==0]['Age'].min()
print("Oldest Person expired on Titanic:",max_age,"\nYoungest Person Expired on Titanic", min_age)
```

```python
# find maximum age of pasenger who servieved
max_age = titanic[titanic['Survived']==1]['Age'].max()
min_age = titanic[titanic['Survived']==1]['Age'].min()
print("Oldest Person Survived on Titanic:",max_age,"\nYoungest Person Survived on Titanic", min_age)
```
